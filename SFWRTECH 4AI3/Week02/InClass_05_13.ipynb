```python
import numpy as np

file = "pima-indians-diabetes.data.csv"
data = np.loadtxt(file, delimiter=",")
```

```python
print("First 5")

print(data[:5])
```

```python
correlation = np.corrcoef(data, rowvar=False)
np.set_printoptions(precision=3, suppress=True)
print("Correlation Matrix")
print(correlation)
```

```python
min_data = np.min(data, axis=0)
max_data = np.max(data, axis=0)

scaled_data = (data - min_data) / (max_data - min_data)
np.set_printoptions(precision=3, suppress=True)

print("first 5 scaled samples")
print(scaled_data[:5])
```

```python
scaled_corr = np.corrcoef(scaled_data, rowvar=False)
np.set_printoptions(precision=3, suppress=True)
print("Correlation matrix of rescaled data:")
print(scaled_corr)
```
