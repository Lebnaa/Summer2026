```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. X and Y matrices needed for the Normal Equation
X_data = np.array([[1, 5],
                   [1, 7],
                   [1, 9],
                   [1, 13],
                   [1, 15],
                   [1, 18]])

Y_data = np.array([[92],
                   [97],
                   [89],
                   [94],
                   [102],
                   [100]])
```

```python
# 2. Using the Normal Equation
theta = np.linalg.inv(X_data.T @ X_data) @ X_data.T @ Y_data
print("theta (intercept, slope):", theta.flatten())
```

```python
# 3. Estimate the output at your input points. 
Y_hat = X_data @ theta
print("Possible outputs:", Y_hat.flatten())
```

```python
# 4. Calculate the Mean Squared Error (MSE) for your predictions. 
mse = np.mean((Y_data - Y_hat) ** 2)
print("MSE:", mse)
```

```python
# 5. Plot
x_vals = X_data[:, 1]
x_line = np.linspace(min(x_vals) - 1, max(x_vals) + 1, 100)
y_line = theta[0] + theta[1] * x_line

plt.figure(figsize=(8, 5))
plt.scatter(x_vals, Y_data, color='blue', marker='o', label='Actual outputs')
plt.scatter(x_vals, Y_hat, color='red', marker='x', s=100, label='Estimated outputs')
plt.plot(x_line, y_line, color='green', linestyle='--', label='Fitted line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Normal Equation - Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
