```python
import numpy as np

file = "pima-indians-diabetes.data.csv"
data = np.loadtxt(file, delimiter=",")
```

```python
print("First 5")
np.set_printoptions(precision=3, suppress=True)
print(data[:5])
```

```python
print("TASK 3: Correlation matrix")
corr_whole = np.corrcoef(data, rowvar=False)
print(corr_whole)
```

```python
print("TASK 4: Rescale to (0,1), print first 5")
data_mix = data.max(axis=0)
data_min = data.min(axis=0)
data_rescaled = (data - data_min) / (data_mix - data_min)

print("TASK 4: First 5 samples after rescaling to (0,1)")
print(data_rescaled[:5])
```

```python
print("TASK 5: Correlation matrix (rescaled data)")
corr_scaled = np.corrcoef(data_rescaled, rowvar=False)
print(corr_scaled)
```

TASK 6: Have correlations changed after rescaling? No, the correlations have not changed after rescaling.
Pearson correlation is scale-invariant. Min-Max scaling is a linear transformation (multiply + add a constant), which preserves the relative distances between values. Therefore all pairwise correlations remain same before and after rescaling.


```python
print("TASK 7: Group sizes ")

class0 = data[data[:, -1] == 0]
class1 = data[data[:, -1] == 1]

print(f"Class 0 (no diabetes): {len(class0)} samples")
print(f"Class 1 (diabetes):    {len(class1)} samples")
```

```python
print("TASK 8: Correlation matrix for class = 0")
corr_class0 = np.corrcoef(class0, rowvar=False)
print(corr_class0)
```

TASK 9: Are class=0 correlations equal to the whole dataset? No, they are not equal. Subsetting to class=0 changes the distribution of samples, so pairwise relationships between features are different from the full dataset. Also, the 'class' column is constant (all 0s) in this group, so its correlation with every other feature is NaN. 

```python
print("TASK 10: Correlation matrix for class = 1")
corr_class1 = np.corrcoef(class1, rowvar=False)
print(corr_class1)
```

TASK 11: Are correlations for class=0 and class=1 equal? No, they are NOT equal. The two groups represent biologically different populations. Internal relationships along features such as glucose, BMI,
insulin, and age are different between diabetic and non-diabetic patients, showing different metabolic patterns. Aggregate correlations across the whole dataset can mask these within-group structures.
